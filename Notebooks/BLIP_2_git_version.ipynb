```python
main_folder_path = '/content/Dataset_simple_sketch'

# get path to all worst images
all_images = []
for root, dirs, files in os.walk(main_folder_path):
    for i in files:
        if i.endswith('.png'):
            all_images.append(os.path.join(root, i))
```

```python
from tqdm import tqdm

# run all of the images through the questioning, add them to a list, save in a csv file
lst_of_worst_probs = []

for i in tqdm(all_images, desc="Processing images: "):
    lst_of_worst_probs.append(questions(image_path=i))
df = pd.DataFrame(lst_of_worst_probs)
df.to_csv('/content/gdrive/MyDrive/BLIP_day_8_given_answers_statment.csv', index=False)
```
