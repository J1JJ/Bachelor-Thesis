# VIT

```python
file_path = '/content/drive/My Drive/lst_of_worst_day_5.csv'
VIT = pd.read_csv(file_path)

VIT
```

```python
probability_columns = VIT.columns[2:]

VIT['max_prob'] = VIT[probability_columns].max(axis=1)
VIT['max_prob_class'] = VIT[probability_columns].idxmax(axis=1)

VIT
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20,20))
x = VIT['Category']
y = VIT["max_prob_class"]

class_to_id = {cls: i for i, cls in enumerate(sorted(set(x)), start=1)}
x_labels_ids = [class_to_id[cls] for cls in x]

plt.figure(figsize=(10,20))
plt.scatter(x_labels_ids, y, alpha=0.3)
plt.xlabel('Actual Class')
plt.ylabel('Predicted Class')
plt.title('Predicted vs True Classes',fontsize=20)
plt.xticks(ticks=sorted(class_to_id.values()), labels=sorted(class_to_id, key=class_to_id.get), rotation=90)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
VIT['correct'] = VIT["max_prob_class"] == VIT['Category']

accuracy = VIT['correct'].mean()
print(f'Accuracy: {accuracy:.2f}')
```

```python
VIT['correct'] = VIT["max_prob_class"] == VIT['Category']

accuracy_by_category_VIT = VIT.groupby('Category')['correct'].mean().reset_index()
accuracy_by_category_VIT.columns = ['Category', 'Accuracy']

print(accuracy_by_category_VIT)
```

```python
def precision_recall_f1(df, category):
    TP = sum((df['Category'] == category) & (df['max_prob_class'] == category))
    FP = sum((df['Category'] != category) & (df['max_prob_class'] == category))
    FN = sum((df['Category'] == category) & (df['max_prob_class'] != category))

    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return precision, recall, f1

results = []
categories = VIT['Category'].unique()
for category in categories:
    precision, recall, f1 = precision_recall_f1(VIT, category)
    results.append({
        'Category': category,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1
    })

results_df = pd.DataFrame(results)
print(results_df)
```

# Llama

```python
file_path = '/content/drive/My Drive/llama_day_6_with_predictions_from_BLIP .csv'
Llama = pd.read_csv(file_path)

Llama
```

```python
Llama["Answer from Llama"]
```

```python
import re

def separate_words(text):
    words_to_separate = ["cellphone", "teddybear", "alarmclock", "santaclaus", "satellitedish", "spaceshuttle"]

    for word in words_to_separate:
        separated_word = ' '.join(list(word))
        text = re.sub(r'\b{}\b'.format(word), separated_word, text)

    return text

Llama['One'] = (Llama["Answer from Llama"]
                .str.replace(r'\*\*(.*?)\*\*', r'\1')
                .str.replace(r'\`(.*?)\`', r'\1')
                .str.replace('**Prediction:** ', ''))
```

```python
Llama
```

```python
plt.figure(figsize=(20,20))
x = Llama["Category"]
y = Llama["One"]

class_to_id = {cls: i for i, cls in enumerate(sorted(set(x)), start=1)}
x_labels_ids = [class_to_id[cls] for cls in x]

plt.figure(figsize=(10,20))
plt.scatter(x_labels_ids, y, alpha=0.3)
plt.xlabel('Actual Class')
plt.ylabel('Predicted Class')
plt.title('Predicted vs True Classes',fontsize=20)
plt.xticks(ticks=sorted(class_to_id.values()), labels=sorted(class_to_id, key=class_to_id.get), rotation=90)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
Llama['correct'] = Llama["One"] == Llama["Category"]

accuracy = Llama['correct'].mean()
print(f'Accuracy: {accuracy:.2f}')
```

```python
Llama['correct'] = Llama["One"] == Llama['Category']

accuracy_by_category_Llama = Llama.groupby('Category')['correct'].mean().reset_index()
accuracy_by_category_Llama.columns = ['Category', 'Accuracy']

print(accuracy_by_category_Llama)
```
